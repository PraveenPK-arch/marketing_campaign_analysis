```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Available matplotlib styles:")
print(plt.style.available)
sns.set_palette('Set2')

url = "https://raw.githubusercontent.com/ArchanaInsights/Datasets/main/marketing_campaign.csv"
df = pd.read_csv(url)

print("First 5 rows of the dataset:")
print(df.head(), "\n")

print(f"Number of rows: {df.shape[0]}, Columns: {df.shape[1]}\n")
print("Dataset Info:")
print(df.info(), "\n")

print("Statistical Summary:")
print(df.describe(), "\n")

# Unique values and counts
print("Unique Campaign IDs:", df['Campaign_ID'].nunique())
print("Unique Locations:", df['Location'].unique())
print("Unique Customer Segments:", df['Customer_Segment'].unique(), "\n")

print("Campaign Type Counts:")
print(df['Campaign_Type'].value_counts(), "\n")

print("Channels Used Counts:")
print(df['Channel_Used'].value_counts(), "\n")

# Exploratory Data Analysis (EDA) and Visualization

# Scatter plot: Acquisition_Cost vs ROI
plt.figure(figsize=(6,4))
sns.scatterplot(x='Acquisition_Cost', y='ROI', data=df)
plt.title("Acquisition Cost vs ROI")
plt.show()

# Bar chart: Avg Conversion_Rate by Channel_Used and Campaign_Type
plt.figure(figsize=(8,5))
sns.barplot(x='Channel_Used', y='Conversion_Rate', hue='Campaign_Type', data=df)
plt.title("Average Conversion Rate by Channel and Campaign Type")
plt.xticks(rotation=45)
plt.show()

# Box plot: Engagement_Score by Campaign_Type
plt.figure(figsize=(6,4))
sns.boxplot(x='Campaign_Type', y='Engagement_Score', data=df)
plt.title("Engagement Score by Campaign Type")
plt.show()

# Bar chart: Average ROI by Company
plt.figure(figsize=(8,5))
roi_by_company = df.groupby('Company')['ROI'].mean().reset_index()
sns.barplot(x='Company', y='ROI', data=roi_by_company)
plt.title("Average ROI by Company")
plt.xticks(rotation=45)
plt.show()

# Heatmap: Correlation between Engagement_Score and Conversion_Rate
plt.figure(figsize=(4,3))
corr_data = df[['Engagement_Score', 'Conversion_Rate']].corr()
sns.heatmap(corr_data, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# Count plot: Target_Audience distribution
plt.figure(figsize=(8,5))
sns.countplot(y='Target_Audience', data=df)
plt.title("Distribution of Target Audience")
plt.show()

# Bar chart: Highest Conversion_Rate by Customer_Segment for each Language
plt.figure(figsize=(10,5))
conv_rate = df.groupby(['Language', 'Customer_Segment'])['Conversion_Rate'].mean().reset_index()
sns.barplot(x='Language', y='Conversion_Rate', hue='Customer_Segment', data=conv_rate)
plt.title("Conversion Rate by Language and Customer Segment")
plt.show()

# Box plot: Acquisition_Cost by Customer_Segment & Channel_Used
plt.figure(figsize=(10,5))
sns.boxplot(x='Customer_Segment', y='Acquisition_Cost', hue='Channel_Used', data=df)
plt.title("Acquisition Cost by Segment and Channel")
plt.xticks(rotation=45)
plt.show()

# Bar chart: Average Conversion_Rate by Language
plt.figure(figsize=(6,4))
lang_conv = df.groupby('Language')['Conversion_Rate'].mean().reset_index()
sns.barplot(x='Language', y='Conversion_Rate', data=lang_conv)
plt.title("Average Conversion Rate by Language")
plt.show()

# Bar chart: Engagement_Score by Channel_Used and Campaign_Type
plt.figure(figsize=(8,5))
sns.barplot(x='Channel_Used', y='Engagement_Score', hue='Campaign_Type', data=df)
plt.title("Engagement Score by Channel and Campaign Type")
plt.xticks(rotation=45)
plt.show()

# Pie chart: Total ROI by Channel_Used
roi_channel = df.groupby('Channel_Used')['ROI'].sum()
plt.figure(figsize=(6,6))
roi_channel.plot.pie(autopct='%1.1f%%')
plt.title("Total ROI by Channel Used")
plt.ylabel('')
plt.show()

# Scatter plot: Clicks vs Impressions for each Channel_Used
plt.figure(figsize=(8,5))
sns.scatterplot(x='Impressions', y='Clicks', hue='Channel_Used', data=df)
plt.title("Clicks vs Impressions by Channel")
plt.show()

# Histogram: Duration distribution
plt.figure(figsize=(6,4))
plt.hist(df['Duration'], bins=10, color='skyblue', edgecolor='black')
plt.title("Campaign Duration Distribution")
plt.xlabel("Days")
plt.ylabel("Frequency")
plt.show()

# Line chart: Conversion_Rate over Date for each Company
plt.figure(figsize=(10,5))
df['Date'] = pd.to_datetime(df['Date'], format='mixed')
sns.lineplot(x='Date', y='Conversion_Rate', hue='Company', data=df)
plt.title("Conversion Rate Over Time by Company")
plt.show()

# Line chart: Engagement_Score over Date
plt.figure(figsize=(10,5))
sns.lineplot(x='Date', y='Engagement_Score', data=df)
plt.title("Engagement Score Over Time")
plt.show()

# Bar chart: Highest Acquisition_Cost by Location
plt.figure(figsize=(8,5))
acq_by_loc = df.groupby('Location')['Acquisition_Cost'].mean().reset_index()
sns.barplot(x='Location', y='Acquisition_Cost', data=acq_by_loc)
plt.title("Average Acquisition Cost by Location")
plt.xticks(rotation=45)
plt.show()

# Bar chart: Conversion_Rate by Location and Target_Audience
plt.figure(figsize=(10,5))
sns.barplot(x='Location', y='Conversion_Rate', hue='Target_Audience', data=df)
plt.title("Conversion Rate by Location and Audience")
plt.xticks(rotation=45)
plt.show()

# Pie chart: ROI by Location
roi_loc = df.groupby('Location')['ROI'].sum()
plt.figure(figsize=(6,6))
roi_loc.plot.pie(autopct='%1.1f%%')
plt.title("ROI by Location")
plt.ylabel('')
plt.show()
```
